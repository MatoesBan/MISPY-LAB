```python
def process_age(df,cut_points,label_names): #funkcja podziału wieku  na określone kategorie
    df["Age"] = df["Age"].fillna(-0.5)
    df["Age_categories"] = pd.cut(df["Age"],cut_points,labels=label_names)
    return df

cut_points = [-1,0, 5, 12, 18, 35, 60, 100]
label_names = ["Missing", 'Infant', "Child", 'Teenager', "Young Adult", 'Adult', 'Senior']

train = process_age(train,cut_points,label_names) #wykonanie funkcji na pliku train
test = process_age(test,cut_points,label_names) #wykonanie funkcji na pliku test
train.head() #wyświetlenie zmian
```

```python
def create_dummies(df,column_name): #funckja tworzxy nową kolumne w której określamy do jakiej klasy należy dany pasażer 
    dummies = pd.get_dummies(df[column_name],prefix=column_name)
    df = pd.concat([df,dummies],axis=1)
    return df

train = create_dummies(train,"Pclass") #wykonanie funkcji na pliku train
test = create_dummies(test,"Pclass") #wykonanie funkcji na pliku test
train.head() #wyświetlenie zmian
```

```python
print("Find most important features relative to target")
corr = train.corr()
corr.sort_values(['Survived'], ascending=False, inplace=True)
corr.Survived #szukamy kolumn które są najbardziej powiązane z kolumną Survived
```

```python
from sklearn.linear_model import LogisticRegression #importujemy biblioteki z modelam który na wyjściu przyjmuje 2 wartości

columns = ['Age', 'Fare', 'Parch', 'Pclass_1', 'Pclass_2'] #wybieramy kolumny na których będizmy dopasowywać obiekt



lr = LogisticRegression() #przypisujemy do zmiennej lr funkcje obiektu typu LogisticRegression
lr.fit(train[columns], train['Survived']) #Doposowywujemy obiekt na podstawie wybranych kolumn i podanych danych o osbach które przeżyły


```

```python
print("Accuracy --> ", lr.score(train[columns], train['Survived'])*100) #wyswietlamy skutecznośc powstałego obiektu
```
